```python
import torch
from ultralytics import YOLO
torch.cuda.empty_cache()
```

```python
# Load a model
res = 864 # full res
model_size = 's'
model = YOLO(f'yolov8{model_size}-pose.pt')
experiment = 'occlusion-v1'

results = model.train(data = 'disengage_videos/data.yaml', 
                    epochs   = 200,
                    batch    = 8,
                    imgsz    = res,
                    device   = 'cuda',
                    workers  = 4,
                    plots    = True,
                    name     = f'{model_size}_{res}_{experiment}',
                    exist_ok = True,
                    augment  = True,
                    fraction = 1.0)
```

```python
results = model.val(data='disengage_videos/data.yaml')
```
